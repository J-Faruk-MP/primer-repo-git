Los operadores relacionales se utilizan para comparar dos valores o expresiones en un programa. Permiten establecer una relacin entre ellos, determinando si una condicin es verdadera o falsa.


Los operadores relacionales ms comunes son:

*Igualdad (==): Compara si dos valores son iguales.

*Desigualdad (!=): Compara si dos valores son diferentes.

*Mayor que (>): Compara si el valor de la izquierda es mayor que el de la derecha.

*Menor que (<): Compara si el valor de la izquierda es menor que el de la derecha.

*Mayor o igual que (>=): Compara si el valor de la izquierda es mayor o igual al de la derecha.

Menor o igual que (<=): Compara si el valor de la izquierda es menor o igual al de la derecha.

Estos operadores son fundamentales en la toma de decisiones dentro de los programas, especialmente en estructuras condicionales y estructuras repetitivas o bucles, donde se evalan condiciones para ejecutar diferentes bloques de cdigo segn el resultado de la comparacin.

# COMPARACIN DE NMEROS

```python
# Definimos las variables a y b
a = 4
b = 3

# Definimos la variable igualdad_a_b y almacenamos en ella la relacin de igualdad entre a y b.
igualdad_a_b = a == b

# igualdad_a_b es un booleano que tendr valor True  si se cumple que a y b son iguales y valor False si dicha condicin no se cumple.
# Imprimimos su valor por pantalla

print(f"Se cumple que {a} y {b} son iguales?")
print(igualdad_a_b)
```

mayor que

```python

# Definimos la variable a_mayor_que_b y almacenamos en ella la relacin correspondiente.
a_mayor_que_b = a > b

# a_mayor_que_b es un booleano que tendr valor True si se cumple que a es mayor que b y valor False si dicha condicin no se cumple.
# Imprimimos su valor por pantalla

print(f"Se cumple que {a} es mayor que {b}?")
print(a_mayor_que_b)
```

menor que

```python
# Definimos la variable a_menor_que_b y almacenamos en ella la relacin correspondiente.
a_menor_que_b = a < b

# a_menor_que_b es un booleano que tendr valor True si se cumple que a es menor que b y valor False si dicha condicin no se cumple.
# Imprimimos su valor por pantalla

print(f"Se cumple que {a} es menor que {b}?")
print(a_menor_que_b)
```

Los operadores relacionales se pueden utilizar tambin para comparar cadenas de caracteres. En este caso, Python ordena las palabras segn el orden en el cdigo ASCII. Veamos algunos casos de uso. Para ello, comencemos definiendo dos variables que usaremos en los ejemplos.

En el caso anterior, Python considera que el string "casa" es menor que el string "perro" ya que la letra "c" se encuentra antes en la numeracin ASCII que la letra "p".

```python
# Definimos la variable c_menor_que_e y almacenamos en ella la relacin correspondiente.
c_menor_que_e = c < e

# c_menor_que_d es un booleano que tendr valor True si se cumple que c es menor que e y valor False si dicha condicin no se cumple.
# Imprimimos su valor por pantalla

print(f"Se cumple que {c} es menor que {e}?")
print(c_menor_que_e)
```

Operadores Logicos

OPERADORES LGICOS
En la programacin, los operadores lgicos son herramientas fundamentales que nos permiten combinar varias condiciones y tomar decisiones basadas en mltiples criterios. Estos operadores se utilizan principalmente en estructuras de control y estructuras repetitivas.

Un operador lgico evala una condicin lgica y devuelve un booleano. Ese booleano adopta valor True si se cumple la condicin y False si no se cumple.

Los operadores lgicos de Python son and, or y not. Veamos uno a uno cmo se comportan.

AND
Se basa en la tabla de verdad de la conjuncin, definida del siguiente modo:


```python
# Definimos las variables a y b
a = 10 == 10
b = 5 == 5

# a y b son verdaderas, por lo tanto a and b es verdadero
print(a and b)
```

```python
# Definimos las variables a y b
a = 10 == 10
b = 5 == 4

# b es falsa, por lo tanto a and b es falso
print(a and b)
```

OR

Se basa en la tabla de verdad de la disyuncin, definida del siguiente modo



El operador or devuelve True si al menos una de las condiciones es verdaderas. Si ambas condiciones son falsas, el resultado ser False.

Veamos algunos ejemplos:

```python
# Definimos las variables a y b
a = 10 == 10
b = 5 == 5

# a y b son verdaderas, por lo tanto a or b es verdadero
print(a or b)
```

```python
# Definimos las variables a y b
a = 10 == 10
b = 5 == 4

# a es verdadera, por lo tanto a or b es verdadero
print(a or b)
```

NOT

Se basa en la tabla de verdad de la negacin, definida del siguiente modo:



El operador not revierte el valor de verdad de un enunciado. Si el enunciado es verdadero, devuelve False y viceversa.

Veamos algunos ejemplos:

```python
# Definimos la variable
a = 10 == 10

# a es verdadera, por lo tanto not a es falso
print(not a)
```

```python
# Definimos la variable
a = 10 == 9

# a es falsa, por lo tanto not a es verdadero
print(not a)
```
