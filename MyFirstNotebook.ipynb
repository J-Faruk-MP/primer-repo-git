# String con formato (f string)

string o cadenas nos permiten construir cadenas de caracteres a partir de literales y variables con una sintaxis mucho mas clara.


```python
nombre = input("Ingrese su nombre: ")
edad = input("Ingrese su edad: ")
# para cadenas uniendo variables con coma no necesitas hacer espacio ,ya que se hace automatico
print("tu numbre es",nombre,"Y tu edad es",edad)
#cuando usas + para concatenar , si hay variables de diferentes tipos de datos , debes convertirlos en string
#con la funcion str()
print("Tu nombre es " + nombre +" y tu edad es " + str(edad))
# es la mas amigable empieza con f"", dento puedes poner llaves para llamar a variables
print(f"Tu nombre es {nombre} y tu edad es {edad}")
```
